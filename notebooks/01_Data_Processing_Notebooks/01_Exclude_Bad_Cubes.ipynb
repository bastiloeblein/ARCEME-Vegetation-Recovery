```python
import pandas as pd
import os
import shutil
```

```python
source_dir = ("../../reports/processing_info_new")
target_dir = ("../../reports/excluded_cubes")
```

```python
bad_cubes = pd.read_csv("../../excluded_cubes.csv")
bad_cubes.head()
```

```python
moved_count = 0
not_found_count = 0

for cube_id in bad_cubes["cube_id"].to_list():
    cube_source_path = os.path.join(source_dir, cube_id)
    cube_target_path = os.path.join(target_dir, cube_id)
    
    if os.path.exists(cube_source_path):
        # Falls der Ordner im Ziel schon existiert, löschen wir ihn dort erst (optional)
        if os.path.exists(cube_target_path):
            shutil.rmtree(cube_target_path)
            
        shutil.move(cube_source_path, cube_target_path)
        moved_count += 1
    else:
        not_found_count += 1

print(f"✅ Fertig! {moved_count} Cubes nach {target_dir} verschoben.")
if not_found_count > 0:
    print(f"⚠️ {not_found_count} Cubes wurden in {source_dir} nicht gefunden (vielleicht noch nicht prozessiert?).")
    
    
```
